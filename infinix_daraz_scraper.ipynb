# Import libraries


```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
```

# Scraping the first page

```python

url = "https://www.daraz.pk/catalog/?q=infinix"
driver = webdriver.Chrome()
driver.get(url)
soup = BeautifulSoup(driver.page_source, "html.parser")
driver.quit()
```

```python
soup.title.text
```

```python
Models=[]
Prices=[]
Links=[]
phone_details={"Models":Models,"Prices(Rs)":Prices,"Buy Now":Links}
```

```python
products = soup.find_all("div", class_="Bm3ON")

for product in products:
    tag_a=(product.find("a"))
    img_tag=tag_a.find("img")
    Models.append(img_tag.attrs["alt"])
```

```python
prices=soup.find_all("div",class_="aBrP0")

for price in prices:
    Prices.append(price.find("span",class_="ooOxS").text.strip("Rs. "))
```

```python
for product in products:
    link=(product.find("a"))
    Links.append("https://"+link.attrs["href"].strip("./"))
```

```python
print(f"There are total {len(products)} Inifinx Phones in first page")
```

```python
import pandas as pd
```

```python
df=pd.DataFrame.from_dict(phone_details)
```

```python
df
```

# Now save the file in with ".csv" ectension

```python
df.to_csv("first_page_infinix_phones.csv")
```

# Now, Scraping first 20 the pages

```python
total_pages=102

Models=[]
Prices=[]
Links=[]
phone_details={"Models":Models,"Prices(Rs)":Prices,"Buy Now":Links}

for i in range(2,23):
    url = f"https://www.daraz.pk/catalog/?page={i}&q=infinix"
    driver = webdriver.Chrome()
    driver.get(url)
    soup = BeautifulSoup(driver.page_source, "html.parser")
    driver.quit()



    products = soup.find_all("div", class_="Bm3ON")
    for product in products:
        tag_a=(product.find("a"))
        img_tag=tag_a.find("img")
        Models.append(img_tag.attrs["alt"])


    price=soup.find_all("div",class_="aBrP0")
    for p in price:
        Prices.append(p.find("span",class_="ooOxS").text.strip("Rs. "))
    for product in products:
        link=(product.find("a"))
        Links.append("https://"+link.attrs["href"].strip("./"))
```

```python
print(f"There are total {len(Models)} Inifinx Phones in 20 page")
```
